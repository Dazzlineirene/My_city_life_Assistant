```python
#Call Asteroids NeoWs API

import requests
base_url = "https://api.nasa.gov/neo/rest/v1/feed"
api_key = "u61DP4EdWB7m45AXTdRtePxiUeqcV42p0Upx067I"  # Replace this with your actual OMDb API key
# Parameters for the GET request
params = {
    "start_date": "2024-01-01",
    "end_date": "2024-01-07",
    "api_key": api_key
}
# Sending the GET request
response = requests.get(base_url, params=params)
data = response.json()

```

```python

asteroids = []

for date in data["near_earth_objects"]:
    for obj in data["near_earth_objects"][date]:
        asteroid_id = obj["id"]
        name = obj["name"]
        min_diameter = obj["estimated_diameter"]["kilometers"]["estimated_diameter_min"]
        absolute_magnitude = obj["absolute_magnitude_h"]
        velocity = float(obj["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"])

        asteroids.append([
            asteroid_id,
            name,
            min_diameter,
            absolute_magnitude,
            velocity
        ])

df = pd.DataFrame(asteroids, columns=[
    "Asteroid ID",
    "Asteroid Name",
    "Min Estimated Diameter (km)",
    "Absolute Magnitude",
    "Relative Velocity (km/s)"
])

df.head()
```

```python
df.to_csv("nasa_asteroids.csv", index=False)
print("CSV file saved as nasa_asteroids.csv")
```
