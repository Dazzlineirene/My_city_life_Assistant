```python
import pandas as pd

url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Create a new table of only high-energy songs
high_energy_songs = df[df['energy'] > 0.9]

# Show how many songs we found
print(f"Found {len(high_energy_songs)} high-energy songs!")

# Show the first 5 tracks and their artists
high_energy_songs[['track_name', 'track_artist', 'energy']].head()
```

```python
# This creates a histogram of the popularity of these high-energy songs
high_energy_songs['track_popularity'].hist(bins=20, color='green', edgecolor='black')
```

```python
popular_songs = df[df['track_popularity'] > 80]
```

```python
popular_songs
```

```python
popular_songs.head(3)
```

```python
popular_songs.tail(2)
```

```python
# Calculate the average happiness of ALL songs
avg_happiness_all = df['valence'].mean()

# Calculate the average happiness of ONLY the hits
avg_happiness_hits = popular_songs['valence'].mean()

print(f"Average Happiness (All Songs): {avg_happiness_all:.2f}")
print(f"Average Happiness (Hit Songs): {avg_happiness_hits:.2f}")
```

```python
# Compare Danceability instead
avg_dance_all = df['danceability'].mean()
avg_dance_hits = popular_songs['danceability'].mean()

print(f"Average Danceability (All): {avg_dance_all:.2f}")
print(f"Average Danceability (Hits): {avg_dance_hits:.2f}")
```

```python
# 1. Group the data by 'playlist_genre'
# 2. Select the 'danceability' column
# 3. Calculate the mean (average) for each group
genre_analysis = df.groupby('playlist_genre')['danceability'].mean()

# Sort them so the highest is at the top
genre_analysis.sort_values(ascending=False)
```

```python
energy_analysis = df.groupby('playlist_subgenre')['energy'].mean()
energy_analysis.sort_values(ascending=False)
```

```python
# 1. We take your analysis and pick the top 10 most energetic subgenres
top_10_energy = energy_analysis.sort_values(ascending=False).head(10)

# 2. We use .plot to make a bar chart
top_10_energy.plot(kind='bar', 
                   figsize=(10, 6), 
                   color='orange', 
                   title='Top 10 High-Energy Subgenres')
```

```python
#Write a program that asks the user for a positive integer n and calculates the sum of all numbers from 1 to n (inclusive).

n = int(input("Enter a number: "))

total = 0
i = 1

while i <= n:
    total += i
    i += 1

print(total)

```

```python
#Write a program that counts how many even numbers are in a list.

x = [1,2,3,4,5,6,7,8]

count = 0

for i in x:
    if i % 2 == 0:
        count += 1
print(count)
        


```

```python
#Write a program that finds the largest number in a list.

y = [10, 45, 5, 8, 14]

largest = y[0]

for i in y:
    if i > largest:
        largest = i

print(largest)
```

```python
#Write a program that counts how many times each word appears in a sentence.

sentence = input("enter a sentence")

words = sentence.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
print(frequency)
```
